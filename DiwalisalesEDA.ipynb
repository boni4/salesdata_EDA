```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
file_name = "/content/Diwali Sales Data.csv"
data = pd.read_csv(file_name, encoding='latin1')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info
```

```python
data.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
data = data.drop_duplicates()
```

```python
data.shape
```

```python
# checking if their are any null values in the data set
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
pd.isnull(data).sum()
```

```python
data.info
```

```python
data['Amount']=data['Amount'].astype(int)
data['Marital_Status'] = data['Marital_Status'].astype('int')
```

```python
data['Gender'] = data['Gender'].replace({"F": "Female", "M": "Male"})
```

```python
data = pd.DataFrame(data)

#change the Marital staus values from 0, 1 to unmarried and married.

data['Marital_Status'] = data['Marital_Status'].replace({0: "Unmarried", 1: "Married"})
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

gender_count = sns.countplot(x='Gender', data=data, palette='viridis')
plt.title("Count of Customers Based on Gender")

# Add labels to the bars
for bars in gender_count.patches:
    gender_count.annotate(format(bars.get_height(), '.0f'),
                          (bars.get_x() + bars.get_width() / 2., bars.get_height()),
                          ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python

# Assuming your data is in a DataFrame named 'data'
sales_gen = data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set the seaborn style
sns.set(style="whitegrid")

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Amount', data=sales_gen, palette='Set1')

# Add labels to the bars
for index, value in enumerate(sales_gen['Amount']):
    plt.text(index, value + 1, f'${value:,.2f}', ha='center', va='bottom', fontsize=10)

# Set plot labels and title
plt.xlabel('Gender')
plt.ylabel('Amount')
plt.title('Total Sales Amount by Gender')

# Show the plot
plt.show()
```

```python
#3.WHICH STATE HAS MORE NUMBER OF ORDERS?

sales_state = data.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(8, 6))
plt.pie(sales_state['Orders'], labels=sales_state['State'], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
plt.title('Total Number of Orders for Top 10 States (Donut Chart)')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
```

```python
# plotting a bar graph for different marital status and see if there is any trend in the data for the married and unmarried

status = data.groupby(['Marital_Status'],as_index = False).sum("Amount").sort_values(by= "Amount", ascending = False)

# plot the data
sns.set(rc={'figure.figsize':(6,6)})
status_plot = sns.barplot(x= 'Marital_Status', y = 'Amount', data = status)
plt.title("Total Revenue Generated by Marital_Status")
plt.ylabel("Revenue generated in 10 millions")

for bars in status_plot.containers:
    status_plot.bar_label(bars)
```

```python
status = data.groupby(['Marital_Status', 'Gender'],as_index = False).sum("Amount").sort_values(by= "Amount", ascending = False)

# plot the data
sns.set(rc={'figure.figsize':(6,6)})
status_plot = sns.barplot(x= 'Marital_Status', y = 'Amount', data = status, hue ='Gender')
plt.title("Total Revenue Generated by Marital_Status")
plt.ylabel("Revenue generated in 10 millions")

for bars in status_plot.containers:
    status_plot.bar_label(bars)
```

```python
occupation = data.groupby(['Occupation'], as_index = False).sum("Amount").sort_values(by = 'Amount', ascending = False)

# plot the data on a bar graph
sns.set(rc={'figure.figsize':(25,6)})
occupation_plot = sns.countplot(data = data, x = 'Occupation')
plt.title("Total Customer Count by Occupation")

for bars in occupation_plot.containers:
    occupation_plot.bar_label(bars)
```

```python
sales_state = data.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(8, 6))
plt.pie(sales_state['Amount'], labels=sales_state['Occupation'], autopct='%1.1f%%', startangle=100)
plt.title('Total Sales Amount by Occupation')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.show()
```

```python
#4.WHICH CATEGORY HAS MORE SOLD PRODUCTS?

sales_state = data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.lineplot(data=sales_state, x='Product_Category', y='Amount', marker='o')
plt.title('Total Sales Amount for Top 10 Product Categories (Line Plot)')
plt.xlabel('Product Category')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
product = data.groupby(['Product_Category'], as_index=False).sum('Amount').sort_values(by = 'Amount', ascending = False).head(5)

#plot the data
sns.set(rc={'figure.figsize':(10,6)})
product_plot = sns.barplot(x = 'Product_Category', y = 'Amount', data = product)
plt.title("TOTAL REVENUE GENERATED BY TOP 5 PRODUCTS")
plt.ylabel("Amount in 10 millions")

for bars in product_plot.containers:
    product_plot.bar_label(bars)
```
